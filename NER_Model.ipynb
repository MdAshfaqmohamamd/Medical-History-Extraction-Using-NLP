```python
import cv2
import pytesseract
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Function to preprocess image using Adaptive Thresholding
def preprocess_image(img: Image.Image) -> np.ndarray:
    gray = cv2.cvtColor(np.array(img), cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    resized = cv2.resize(gray, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)  # Upscale image
    processed_image = cv2.adaptiveThreshold(
        resized, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 63, 12
    )  # Apply adaptive thresholding
    return processed_image

# Load the image
image_path = r"C:\Users\Admin\Desktop\DATASET_mp2\Dataset\prescription_dataset\printed\PRS208C4021562c.jpg"  # Replace with your image path
original_image = Image.open(image_path)

# Display the original image
plt.figure(figsize=(8, 6))
plt.imshow(original_image)
plt.axis("off")
plt.title("Original Prescription Image")
plt.show()

# Preprocess the image
processed_image = preprocess_image(original_image)

# Display the processed image
plt.figure(figsize=(8, 6))
plt.imshow(processed_image, cmap="gray")
plt.axis("off")
plt.title("Preprocessed Image for OCR")
plt.show()

# Use Tesseract OCR to extract text
text = pytesseract.image_to_string(processed_image, lang="eng")

# Print extracted text
print("Extracted Text:")
print(text)
```

```python
import re
import spacy
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download NLTK stopwords if not already downloaded
nltk.download('punkt')
nltk.download('stopwords')

# Load Spacy's English model (replace with a medical-specific model if needed)
nlp = spacy.load("en_core_web_sm")  # Use "en_core_sci_md" for SciSpacy medical model

# Function to clean text (remove special characters, extra spaces, and convert to lowercase)
def clean_text(text: str) -> str:
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W+', ' ', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Function for NLP processing (Tokenization, Stopword Removal, NER)
def process_text(text: str):
    # Cleaning text
    cleaned_text = clean_text(text)
    
    # Tokenization
    tokens = word_tokenize(cleaned_text)
    
    # Stopword Removal
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    # Named Entity Recognition (NER) using Spacy
    doc = nlp(cleaned_text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]

    return {
        "cleaned_text": cleaned_text,
        "tokens": filtered_tokens,
        "entities": entities
    }



# Perform NLP processing
nlp_results = process_text(text)

# Display NLP results
print("\nCleaned Text:\n", nlp_results["cleaned_text"])
print("\nTokens:\n", nlp_results["tokens"])
print("\nNamed Entities:\n", nlp_results["entities"])
```
